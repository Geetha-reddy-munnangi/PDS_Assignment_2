#a) Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.

```python
import pandas as pd
import numpy as np
```

```python
#loading the dataset
df = pd.read_csv("/content/train.csv")
df
```

```python
df.shape # dataset shape
```

```python
df.info()
```

```python
df.isnull().sum()  #checking for the null values
```

```python
missing_cols = ['Mileage', 'Engine', 'Power', 'Seats']

for col in missing_cols:
    if df[col].dtype == 'object':
        # For categorical columns using mode
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # For numeric columns, using mean
        df[col].fillna(df[col].mean(), inplace=True)
#removing the 'New_Price' because more than 80% of data is missing so i  dropping the column
df.drop('New_Price', axis=1, inplace=True)

print("Missing values handled successfully.")
```

#C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value

```python
df_2 = pd.read_csv("/content/cleaned_data_train_B.csv")
df_2
```

```python
# Removing extra spaces and making sure the format matches mapping
df_2['Transmission'] = df_2['Transmission'].str.strip()
df_2['Fuel_Type'] = df_2['Fuel_Type'].str.strip()

# Converting values to 0 and 1
df_2['Transmission'] = df_2['Transmission'].map({'Automatic': 0, 'Manual': 1})
df_2['Fuel_Type'] = df_2['Fuel_Type'].map({'Petrol': 0, 'Diesel': 1})
```

#e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

```python
df_4 = pd.read_csv("/content/cleaned_data_train_D.csv")
df_4
```

```python

selected_columns = df_4[['Name', 'Year', 'Fuel_Type', 'Transmission', 'Price']]
filtered_data = df_4[df_4['Year'] > 2015]
renamed_df_4 = df_4.rename(columns={'Mileage': 'Fuel_Efficiency', 'Power': 'Engine_Power'})
df_4['Age'] = 2024 - df_4['Year']
sorted_df_4 = df_4.sort_values(by='Price', ascending=False)
summary_stats = df_4.groupby('Fuel_Type')['Price'].mean()
summary_stats_complex = df_4.groupby('Fuel_Type').agg({'Price': ['mean', 'min', 'max']})

print("Selected Columns:")
print(selected_columns.head())

print("\nFiltered Data (Year > 2015):")
print(filtered_data.head())

print("\nRenamed Columns:")
print(renamed_df_4.head())

print("\nNew Column 'Age':")
print(df_4[['Name', 'Year', 'Age']].head())

print("\nSorted by Price:")
print(sorted_df_4.head())

print("\nSummary Statistics (Mean Price by Fuel Type):")
print(summary_stats)

print("\nSummary Statistics (Complex):")
print(summary_stats_complex)
```
